## Part 1: Data Collection
In this notebook, some basic data cleaning is done in order to isolate the data to only the solo matches that the project is focused on. Also, to limit the scale of the project, the data of only 1,000 matches is taken.

```python
# Import libraries
import pandas as pd
```

```python
# import the matches_stats database
match_stats = pd.read_csv('../datasets/matches_stats.csv') 
display(match_stats.shape)
match_stats.head()
```

<a id='isolate'></a>
### 1.  Isolating solo matches
Since there are various form of match modes, there is no clear manner to differentiate the match modes other than the number of players from the `teamNames` column. As such, we will be isolating the names first, and then from there, isolate the solo matches that we require.

```python
## Isolating solo matches
solos = []
names_list = []

for i in match_stats.index:
    # Splitting the names
    name_dict = match_stats['teamNames'][i][1:-1].split(", ")
    
    # Creating a list of names for each row, with the account id and name split
    name_list = []
    for j in name_dict:
        name = j.split(': ')
        name_list.extend(name)
    
    # isolating the actual name
    names = []
    for k in range(len(name_list)):
        if k % 2 != 0:
            name = name_list[k][1:-1]
            names.append(name)
    
    # Adding to list of names
    names_list.append(names)
    
    # Adding to list that differentiate solo matches
    if len(name_dict) == 1:
        solos.append(1)
    else:
        solos.append(0)    

# Creating the two new columns of solo matches and names        
match_stats['solos'] = solos
match_stats['names'] = names_list
```

```python
# Looking at the dataframe again
match_stats
```

```python
# Removing the brackets from the names
names_list = []
df = match_stats.copy()
for i in df.index:
    name = df.loc[i, 'names']
    name2 = str(name)[2:-2]
    names_list.append(name2)
df['names'] = names_list
```

```python
# Isolating the solo matches and passing into a new dataframe
match_stats_solo = df[df['solos'] == 1]
match_stats_solo.reset_index(drop = True, inplace = True)
```

```python
# Checking the new dataframe
match_stats_solo
```

<a id='narrow'></a>
### 2.  Narrow the scope

Looking at the above dataframe, we have 464,346 rows of data to go through. That means 464,346 player statistics that we will need to extract, though this includes all duplicates and other possible issues, which meant too much time will be needed to extract the data we require. 

As such, there is a need to limit the number of matches in order to make this process more feasible.

```python
# Extracting the session id for solo matches
sessionid_list = []
for session in match_stats_solo['sessionId']:
    if session not in sessionid_list:
        sessionid_list.append(session)
display(len(sessionid_list))
sessionid_list
```

```python
# Extracting the rows for the first 1000 sessions
match_stats_solo1 = match_stats_solo.loc[match_stats_solo['sessionId'].isin(sessionid_list[:1000])]
match_stats_solo1
```

```python
# Saving to file
match_stats_solo1.to_csv('../datasets/matches_stats_solo.csv', index = False)
```
