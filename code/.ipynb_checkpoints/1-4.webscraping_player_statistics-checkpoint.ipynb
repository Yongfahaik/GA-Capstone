## Part 1: Data Collection
In this notebook, web scraping is conducted to gather the player statistics based on the player names gathered from above.

```python
# Import libraries
import requests
import pandas as pd
import json

from requests.exceptions import ConnectionError
from http.client import RemoteDisconnected
from http.client import HTTPException
from urllib3.exceptions import ProtocolError
```

```python
# API key for scraping using the TRN API
headers = {'TRN-Api-Key' : '9251af32-89bf-4414-a81a-83eb9bf3c29e', 
          'User-Agent': 'Mozilla/5.0'}
```

```python
# Setting the url for the API
base_url = 'https://api.fortnitetracker.com/v1/profile/'
platform = 'all'
```

```python
# Import the matches_stats_solo database
match_df = pd.read_csv('../datasets/matches_stats_solo.csv') 
display(match_df.shape)
match_df.head()
```

<a id='scrape'></a>
### 1. Scraping Player Stats

```python
# Extracting the list of names
names_list = []
for name in match_df['names']:
    names_list.append(name)
names_list[:5]
```

```python
# Creating a set to remove duplicate names
names_set = list(set(names_list))
display(len(names_set))
names_set
```

```python
%%time
# Scraping the player statistics
player_stats = []
json_errors = []
http_errors = []
for name in names_set:
    print(f'Scraping player {name}')
    try:
        player_url = base_url + platform + '/' + name
        res = requests.get(player_url, headers=headers)
        result = res.json()

        player = {}
        player['name'] = result['epicUserHandle']
        player['accountId'] = result['accountId']

        for i in result['lifeTimeStats']:
            player_str = 'lifetime' + '_' + i['key']
            player[player_str] = i['value']

        for i in result['stats']['p2']:
            player_str2 = 'solo' + '_' + i
            player[player_str2] = result['stats']['p2'][i]['value']

        for i in result['stats']['p10']:
            player_str3 = 'duos' + '_' + i
            player[player_str3] = result['stats']['p10'][i]['value']

        for i in result['stats']['p9']:
            player_str4 = 'squads' + '_' + i
            player[player_str4] = result['stats']['p9'][i]['value']
    
        for i in result['stats']['trios']:
            player_str5 = 'trios' + '_' + i
            player[player_str5] = result['stats']['trios'][i]['value']
    
        player_stats.append(player)
        
    except json.decoder.JSONDecodeError:
        json_errors.append(name)
        print("String could not be converted to JSON")
        continue
    
    except KeyError:
        player = {}
        player['name'] = name
        player_stats.append(player)
        print ('Error in scraping, skipping.')
        continue
        
    except (ConnectionError, RemoteDisconnected, ProtocolError, HTTPException) as e:
        http_errors.append(name)
        print("Connection errors. Skipping and retry later.")
```

```python
# Passing to a dataframe
players_df = pd.DataFrame.from_dict(player_stats)
display(players_df.shape)
players_df.head()
```

```python
# Saving to file
players_df.to_csv('../datasets/players_stats.csv', index = False)
```

```python
# Saving to file
json_df = pd.DataFrame(json_errors, columns = ['names'])
json_df.to_csv('../datasets/json_errors.csv', index = False)
```

```python
# Saving to file
names_df = pd.DataFrame(names_set, columns=['names'])
names_df.to_csv('../datasets/players_names.csv', index = False)
```
