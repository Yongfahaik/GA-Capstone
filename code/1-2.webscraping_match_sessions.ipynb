## Part 1: Data Collection
In this notebook, web scraping is conducted based on the `match ids` that we have obtained previously to gather the match statistics data that we will need for modelling.

<a id='imports'></a>
### 1. Imports (All imported libraries are added here)

```python
# Import libaries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup

import selenium
from selenium import webdriver
```

<a id='scrape'></a>
### 2.  Web scraping

```python
# Defining base url
base_url = "https://fortnitetracker.com"
```

```python
# Importing the match ids for web scraping
match_ids = pd.read_csv('../datasets/match_ids.csv')
display(match_ids.shape)
match_ids.head()
```

<a id='sessions'></a>
### 2.1 Event Sessions

At this point in time, using the `match_ids`, we will be scraping for the data of the event sessions which are the match statistics that we require.

```python
%%time
# Scraping the event sessions
matches_stats = []
driver = webdriver.Chrome()
for j in match_ids.index:
        match_stats = {}
        print(f'Scraping match_id {j}')
        try:
            driver.implicit_wait(1) # Time delay of 1 sec
            driver.get(base_url + '/events/sessions/' + match_ids['match_id'][j])
            imp_sessionData = driver.execute_script('return imp_sessionData')
            matches_stats.extend(imp_sessionData)
        except Exception:
            print ('Error in scraping, skipping.')
            continue
driver.quit()
```

```python
# Putting the data to a dataframe and checking its shape
matches_df = pd.DataFrame(matches_stats)
display(matches_df.shape)
matches_df.head()
```

```python
# Saving to file
matches_df.to_csv('../datasets/matches_stats.csv', index = False)
```
